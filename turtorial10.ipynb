##urllib

```python
import urllib.request
request_url = urllib.request.urlopen('https://irabrod.ir')
print(request_url.read())
```

```python
import cv2
from matplotlib import pyplot as plt
url = "https://irabrod.ir/wp-content/uploads/2023/04/interstellar-768x512.jpg"
urllib.request.urlretrieve(url, "interestellar.jpg")

# Opening the image and displaying it (to confirm its presence)
img = cv2.imread("interestellar.jpg")#Image.open(r"interestellar.jpg")
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

##requests

```python
import requests
METHOD = 'GET'
"""
GET => to get data
POST = > TO send data
PUT = > to update/save data
DELETE = > to delete data
"""
payload = {
    "key1":"val1"
}
headers = {
    "Authorization":"!@#$%^&"
}
x = requests.request(METHOD,'https://httpbin.org/get',
                     data=payload,
                     headers=headers
                     )
```

```python
x.json()
```

```python
import requests
METHOD = 'POST'
"""
GET => to get data
POST = > TO send data
PUT = > to update/save data
DELETE = > to delete data
"""
payload = {
    "key1":"val1"
}
headers = {
    "Authorization":"!@#$%^&"
}
x = requests.request(METHOD,'https://postman-echo.com/post',
                     data=payload,
                     headers=headers
                     )
```

```python
x.json()
```
